##### post request to theverge.com


```python
base_Link = "https://theverge.com"
Link_text = requests.get(base_Link).text
print(Link_text)
```

```python
soup = bs(Link_text, 'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

##### After inspecting the website found that there is an ordered list consist of items which have news heading and other required details

```python
headlines_list = soup.find('ol')
```

```python
headlines = headlines_list.find_all('li')
```

```python
for headline in headlines:
    print(headline.__dict__)
```

```python
output = []
```

### Logic to loop among the list times and fetch the reqired details such as url, headline, authors and date.

```python
for i in range(len(headlines)):
    h2 = headlines[i].find('h2')
    link2 = h2.find('a').get('href')
    s= headlines[i].find('div',class_ = 'inline-block').find_all('a')
    author =  " and ".join([i.text for i in s])
    # print(author)
    date = headlines[i].find_all('div')[-1].text
    temp_dict = {'id':i,"URL":base_Link+link2,"headline":h2.text , "author":author , "date":date}
    output.append(temp_dict)
```

#### The output array which have dictionaries of all the details
